# General

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16, 16)

image = cv2.imread("/home/enacom/TestesVisao/SOURCES/pokemon3d.jpg")
plt.imshow(image)
```

# sepFilter2D

<font size="4"> Applies a separable linear filter to an image.
<br> <br>
The function applies a separable linear filter to the image. That is, first, every row of src is filtered with the 1D kernel kernelX. Then, every column of the result is filtered with the 1D kernel kernelY. The final result shifted by delta is stored in dst .
</font>

<font size="4"> **dst	=	cv.sepFilter2D(	src, ddepth, kernelX, kernelY[, dst[, anchor[, delta[, borderType]]]]	)**</font>


## Using Kernel 1D

```python
dst = cv2.sepFilter2D(image, -1, kernelX=1, kernelY=5 , dst=None, anchor=(-1, -1), borderType=cv2.BORDER_CONSTANT)
plt.imshow(dst)
```

# Using Kernel 2D

```python
dst = cv2.sepFilter2D(image, -1, kernelX=(1, 1), kernelY=(5, 5) , dst=None, anchor=(-1, -1), borderType=cv2.BORDER_CONSTANT)
plt.imshow(dst)
```

# Changing Delta value

```python
dst_ori = cv2.sepFilter2D(image, -1, kernelX=1, kernelY=5 , dst=None, anchor=(-1, -1), borderType=cv2.BORDER_CONSTANT)

dst = cv2.sepFilter2D(image, -1, kernelX=1, kernelY=5, dst=None, anchor=(-1, -1), delta=100, borderType=cv2.BORDER_CONSTANT)
dst2 = cv2.sepFilter2D(image, -1, kernelX=1, kernelY=5, dst=None, anchor=(-1, -1), delta=-100, borderType=cv2.BORDER_CONSTANT)

fig, axs = plt.subplots(1, 3)
axs[0].imshow(dst_ori)
axs[0].axis("off")
axs[0].set_title("after 1d kernel")
axs[1].imshow(dst)
axs[1].axis("off")
axs[1].set_title("delta 100")
axs[2].imshow(dst2)
axs[2].axis("off")
axs[2].set_title("delta -100")
```
