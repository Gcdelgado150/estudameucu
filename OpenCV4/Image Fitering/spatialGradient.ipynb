# spatialGradient

Calculates the first order image derivative in both x and y using a Sobel operator.

Equivalent to calling:
    
Sobel( src, dx, CV_16SC1, 1, 0, 3 ); <br>
Sobel( src, dy, CV_16SC1, 0, 1, 3 );

<font size="4"> **	dx, dy	=	cv.spatialGradient(	src[, dx[, dy[, ksize[, borderType]]]]	)**</font>

<font size="4"> **dst = cv.Sobel( src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]] )**</font>

```python
dx, dy = cv2.spatialGradient(image[:,:,0])
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(dx)
axs[1].imshow(dy)
plt.show()
```

```python
dx = cv2.Sobel(image[:,:,0], ddepth=-1, dx=0, dy=1, dst=None, ksize=1)
plt.imshow(dx)
```
