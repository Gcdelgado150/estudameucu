# Median Blur

Blurs an image using the median filter.
<br><br>
The function smoothes an image using the median filter with the ksize√óksize aperture. 
<br>Each channel of a multi-channel image is processed independently. In-place operation is supported.

$$\begin{bmatrix} 1 & 1 & 1 \\ 3 & 3 & 3 \\ 1 & 1 & 1 \end{bmatrix} =>> \begin{bmatrix} . & . & . \\ . & 1.667 & . \\ . & . & . \end{bmatrix}$$

<font size="4"> **dst	=	cv.medianBlur(	src, ksize[, dst]	)**</font>

```python
dst1 = cv2.medianBlur(image, 3)
dst2 = cv2.medianBlur(image, 5)
dst3 = cv2.medianBlur(image, 7)
dst4 = cv2.medianBlur(image, 9)
dst5 = cv2.medianBlur(image, 13)
dst6 = cv2.medianBlur(image, 17)

fig, axs = plt.subplots(2, 3)
axs[0][0].imshow(dst1)
axs[0][1].imshow(dst2)
axs[0][2].imshow(dst3)
axs[1][0].imshow(dst4)
axs[1][1].imshow(dst5)
axs[1][2].imshow(dst6)
plt.show()
```
