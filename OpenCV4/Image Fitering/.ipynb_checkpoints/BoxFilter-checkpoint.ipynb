<font size="5">
Blurs an image using the box filter.

The function smooths an image using the kernel:

<br><br>
$$K = \alpha\begin{bmatrix} 1 & 1 & \dots & 1 \\ 1 & 1 & \dots & 1 \\ \dots &  &  &  \\ 1 & 1 & \dots & 1 \end{bmatrix},$$

where

$$ \alpha = \frac{1}{ksize_{width} * ksize_{height}} $$ when normalize = True
$$ \alpha = 1 $$ when normalize = False
</font>

	dst	=	cv.boxFilter(	src, ddepth, ksize[, dst[, anchor[, normalize[, borderType]]]]	)

# Defines

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/home/enacom/TestesVisao/SOURCES/pokemon3d.jpg")[:, :, 0]
```

# Regular boxFilter

```python
box = cv2.boxFilter(image, ddepth=-1, ksize=(3, 3), dst=None, anchor=(-1, -1), normalize=True, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box)
```

# Increased kernel size

```python
box2 = cv2.boxFilter(image, ddepth=-1, ksize=(7, 7), dst=None, anchor=(-1, -1), normalize=True, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box2)
```

# Change anchor

```python
box3 = cv2.boxFilter(image, ddepth=-1, ksize=(7, 7), dst=None, anchor=(2, 2), normalize=True, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box3)
```

# Normalize

```python
box4 = cv2.boxFilter(image, ddepth=-1, ksize=(3, 3), dst=None, anchor=(-1, -1), normalize=False, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box4)
```

<font size="6">

```python
box5 = cv2.boxFilter(image, ddepth=-1, ksize=(5, 5), dst=None, anchor=(-1, -1), normalize=False, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box5)
```

<font size="5">
    So what happens here is that normalize Flag is indicating whether or not the kernel applied to the image is normalized by its size, or not.
    <br> 
    <br> 
    A kernel with sum == 1 doesnt change the mean value of the image (as seen above)
    <br> 
    <br>
    Meanwhile, with sum > 1 the mean value will change as higher as the size of kernel is
    <br>
    <br>
    So, probably, after applying a very large sized kernel, say (31, 31), our image will be full white (255).
</font>

```python
box6 = cv2.boxFilter(image, ddepth=-1, ksize=(31, 31), dst=None, anchor=(-1, -1), normalize=False, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box6)
```

```python
np.mean(box6)
```

<font size="5">
    Because each iteration of the convolution will produce a massive 31 sum and:
    <br>
    $$\frac{255}{31*31} = 0.265,$$
<br>
Meaning that: For a new mapped value not be 255 the mean value of the box [31, 31] in image can't be greater than 0.265. Which is really unrealistic. (Because the original image will be all black for that).
</font>

# Lets try that

If each pixel in the image values 0.1, the value mapped will be:

$$31*31*0.1 = 96$$

This will form a gray image

```python
image_black = np.ones(image.shape)/10
plt.imshow(image_black)
```

```python
box7 = cv2.boxFilter(image_black, ddepth=-1, ksize=(31, 31), dst=None, anchor=(-1, -1), normalize=False, borderType=cv2.BORDER_CONSTANT)
plt.imshow(box7)
```
