# Bilateral filter

BilateralFilter can reduce unwanted noise very well while keeping edges fairly sharp. However, it is very slow compared to most filters.

```python
ds = [0.01, 0.1, 0.3, 1, 3, 9]
sigmas1 = [1, 30, 90, 150, 180, 300]
sigmas2 = [1, 30, 90, 150, 180, 300]
borders = [cv2.BORDER_CONSTANT, cv2.BORDER_REPLICATE, cv2.BORDER_REFLECT, cv2.BORDER_WRAP, cv2.BORDER_REFLECT_101, cv2.BORDER_TRANSPARENT, cv2.BORDER_ISOLATED]

param_grid = {'ds': ds, 's1': sigmas1, 's2': sigmas2, 'borders': borders}

for kds in ds:
    for ks1 in sigmas1:
        for ks2 in sigmas2:
            for kb in borders:
                filtered = cv2.bilateralFilter(image, d=1,sigmaColor=150,sigmaSpace=150)
                plt.imshow(filtered)
```
