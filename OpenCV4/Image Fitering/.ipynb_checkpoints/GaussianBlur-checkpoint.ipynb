<font size="3"> 
Blurs an image using a Gaussian filter.
<br><br>
The function convolves the source image with the specified Gaussian kernel. In-place filtering is supported.

</font>

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/home/enacom/TestesVisao/SOURCES/pokemon3d.jpg")
def get_gaussian_blur(image, k_size=k_size, sigmaX=sigmaX, sigmaY=sigmaY, borderType=cv2.BORDER_CONSTANT):
    """sigmaX: Gaussian kernel standard deviation in X direction.
       sigmaY: Gaussian kernel standard deviation in Y direction.
    """
    
    img = cv2.GaussianBlur(image, (k_size, k_size), sigmaX, None, sigmaY, borderType=borderType)
    plt.imshow(img)
```

```python
k_size = 15
sigmaX = 0.2 # Gaussian kernel standard deviation in X direction.
sigmaY = 0.5 # Gaussian kernel standard deviation in Y direction.
```
