# Defines

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
```

```python
elements = [cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE]
k_size = 5
it = 3
```

```python
def get_dilation(image, k_size=3, n_iter=3, element=0):
    kernel = cv2.getStructuringElement(elements[element], (k_size, k_size))
    img = cv2.dilate(image, None, kernel, iterations = n_iter)
    plt.imshow(img)
```
