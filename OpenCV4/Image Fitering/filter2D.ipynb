```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
```

```python
elements = [cv2.MORPH_RECT, cv2.MORPH_CROSS, cv2.MORPH_ELLIPSE]
```

```python
image = cv2.imread("/home/enacom/TestesVisao/SOURCES/pokemon3d.jpg")
```

```python
k_size = 3
kernel = cv2.getStructuringElement(elements[0], (k_size, k_size))
dst = cv2.filter2D(image, -1, kernel, None,)
plt.imshow(dst)
```
