# Laplacian

Calculates the Laplacian of an image.

The function calculates the Laplacian of the source image by adding up the second x and y derivatives calculated using the Sobel operator:

$$dst= Δsrc= \frac{∂^2src}{∂x^2} + \frac{∂^2src}{∂y^2}$$
This is done when ksize > 1. When ksize == 1, the Laplacian is computed by filtering the image with the following 3×3 aperture:

$$\begin{bmatrix} 0 & 1 & 0 \\ 1 & -4 & 1 \\ 1 & 1 & 0 \end{bmatrix},$$

<font size="4">**dst	=	cv.Laplacian(	src, ddepth[, dst[, ksize[, scale[, delta[, borderType]]]]]	)**</font>


```python
dst_0 = cv2.Laplacian(image, -1, None, ksize=3)
dst_1 = cv2.Laplacian(image, -1, None, ksize=3, scale=2)
dst_2 = cv2.Laplacian(image, -1, None, ksize=3, scale=4)

fig, axs = plt.subplots(1, 3)
axs[0].imshow(dst_0)
axs[0].axis('off')
axs[1].imshow(dst_1)
axs[1].axis('off')
axs[2].imshow(dst_2)
axs[2].axis('off')
plt.show()
```
