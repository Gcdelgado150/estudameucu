# Bilateral filter

BilateralFilter can reduce unwanted noise very well while keeping edges fairly sharp. However, it is very slow compared to most filters.

```python
ds = [1, 3, 5]
sigmas1 = [90, 150]
sigmas2 = [90, 150]
borders = [cv2.BORDER_CONSTANT]

for kds in ds:
    for ks1 in sigmas1:
        for ks2 in sigmas2:
            for kb in borders:
                filtered = cv2.bilateralFilter(image, d=kds,sigmaColor=ks1,sigmaSpace=ks2, borderType=kb)
                plt.imshow(filtered)
                plt.title("{}-{}-{}-{}".format(kds, ks1, ks2, kb))
                plt.show()
                plt.close()
```
