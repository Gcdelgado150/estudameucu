# General

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import cv2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16, 16)

filename="/home/enacom/TestesVisao/SOURCES/pokemon3d.jpg"
```

# imread

```python
image = cv2.imread(filename)

print(image.shape, image.dtype)
# Test if they are equal
fig, axs = plt.subplots(1, 3)
axs[0].imshow(image[:,:,0])
axs[1].imshow(image[:,:,1])
axs[2].imshow(image[:,:,2])
plt.show()
```

## Flags

### Reading as UNCHANGED
Will consider 4-alpha-channel

```python
image_unchanged = cv2.imread(filename, flags=cv2.IMREAD_UNCHANGED)

print(image_unchanged.shape, image_unchanged.dtype)
# Test if they are equal
fig, axs = plt.subplots(1, 3)
axs[0].imshow(image_unchanged[:,:,0])
axs[1].imshow(image_unchanged[:,:,1])
axs[2].imshow(image_unchanged[:,:,2])
plt.show()
```

### Reading in grayscale
Will convert it from BGR to GRAY. **IT MAY BE DIFFERENT FROM cvtColor**

```python
image_gray = cv2.imread(filename, flags=cv2.IMREAD_GRAYSCALE)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(image_gray.shape, image_gray.dtype)
# Test if they are equal
fig, axs = plt.subplots(1, 2)
axs[0].imshow(gray)
axs[1].imshow(image_gray)
plt.show()
```

### Reading in color
Will convert image to BGR

```python
img_color = cv2.imread(filename, flags=cv2.IMREAD_COLOR)

print(img_color.shape, img_color.dtype)
# Test if they are equals
fig, axs = plt.subplots(1, 3)
axs[0].imshow(img_color[:,:,0])
axs[1].imshow(img_color[:,:,1])
axs[2].imshow(img_color[:,:,2])
plt.show()
```

### Reading in ANYDEPTH
If set, the image is read in any possible color format.

```python
img_color = cv2.imread(filename, flags=cv2.IMREAD_ANYDEPTH)
print(img_color.shape, img_color.dtype)

# Test if they are equals
fig, axs = plt.subplots(1, 2)
axs[1].imshow(image[:,:,0])
axs[0].imshow(img_color)
plt.show()
```

# haveImageReader
Returns true if the specified image can be decoded by OpenCV.

```python
cv2.haveImageReader(filename)
```

# haveImageWriter
Returns true if an image with the specified filename can be encoded by OpenCV.

```python
cv2.haveImageWriter(filename)
```

# imencode

```python
_, outp_buffer = cv2.imencode(".tiff", image)
```

```python
outp_buffer
```

# imdecode

```python
image_decoded = cv2.imdecode(outp_buffer, 1)
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(image)
axs[1].imshow(image_decoded)
plt.show()
```
