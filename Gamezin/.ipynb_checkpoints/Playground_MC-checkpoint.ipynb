```python
import numpy as np

# Beale's function
def objective(X):
    x = X[0]; y = X[1]
    return (1.5 - x + x*y)**2 + (2.25 - x + x*y**2)**2 + (2.625 - x + x*y**3)**2
```

```python
X = []
Y = []
Z = []

for i in range(5):
    for j in range(5):
        V = (i, j)
        X.append(i)
        Y.append(j)
        Z.append(objective(V))
```

```python
# function boundaries
xmin, xmax, xstep = -4.5, 4.5, .9
ymin, ymax, ystep = -4.5, 4.5, .9

# Let's create some points
x1, y1 = np.meshgrid(np.arange(xmin, xmax + xstep, xstep), np.arange(ymin, ymax + ystep, ystep))
```

```python
# initial guess
x0 = [4., 4.]  
f0 = objective(x0)
print (f0)
```

```python
from scipy.optimize import minimize
bnds = ((xmin, xmax), (ymin, ymax))
minimum = minimize(objective, x0, bounds=bnds)
print(minimum)
```

```python
import tensorflow as tf
import keras
from keras import layers
from keras import models
from keras import utils
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import Activation
from keras.regularizers import l2
from keras.optimizers import SGD
from keras.optimizers import RMSprop
from keras import datasets

from keras.callbacks import LearningRateScheduler
from keras.callbacks import History

from keras import losses
from sklearn.utils import shuffle

np.random.seed(5)
```

```python
mnist = keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train.shape, y_train.shape
```

```python
# normalize the data
x_train, x_test = x_train/255.0, x_test/255.0

# reshape the data into 1D vectors
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

num_classes = 10
# Convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
epochs=60
learning_rate = 0.1
decay_rate = learning_rate / epochs
momentum = 0.8
sgd = SGD(lr=learning_rate, momentum=momentum, decay=decay_rate, nesterov=False)
```

```python
def get_model(input_dim):
    lr_model = Sequential()
    lr_model.add(Dense(64, activation=tf.nn.relu, kernel_initializer='uniform', 
                    input_dim = input_dim)) 
    lr_model.add(Dropout(0.1))
    lr_model.add(Dense(64, kernel_initializer='uniform', activation=tf.nn.relu))
    lr_model.add(Dense(num_classes, kernel_initializer='uniform', activation=tf.nn.softmax))

    return lr_model

input_dim = x_train.shape[1]
model = get_model(input_dim)
# compile the model
model.compile(loss='categorical_crossentropy', optimizer=sgd,
              metrics=['acc'])
```

```python
# Fit the model
batch_size = int(input_dim/100)

lr_model_history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
```

# Callback to a learning rate changing by exponential

```python
model = get_model(input_dim)
# compile the model
model.compile(loss='categorical_crossentropy', optimizer=sgd,
              metrics=['acc'])

# define the learning rate change 
def exp_decay(epoch):
    lrate = learning_rate * np.exp(-decay_rate*epoch)
    return lrate
    
# learning schedule callback
loss_history = History()
lr_rate = LearningRateScheduler(exp_decay)
callbacks_list = [loss_history, lr_rate]

# you invoke the LearningRateScheduler during the .fit() phase
exponential_decay_model_history = model.fit(x_train, y_train,
                                            batch_size=batch_size,
                                            epochs=epochs,
                                            callbacks=callbacks_list,
                                            verbose=1,
                                            validation_data=(x_test, y_test))
```

# Changing Loss function

```python
# defining the parameters for RMSprop (I used the keras defaults here)
rms = RMSprop(lr=0.001, rho=0.9, epsilon=None, decay=0.0)
```

```python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer=rms,
              metrics=['acc'])
```
