```python
path = "screenshot/3/"
import os

STARTY = 5
GRID = 100
LEFT = 173
RIGHT = LEFT + 170

def get_sum(stream):
    return "Tree" if sum(stream) == 4170 else "Nada"

print_count = 0
print_limit = 50

for p in os.listdir(path):
    img_right = []
    img_left = []
    title_left = []
    title_right = []
    
    img_orig = cv2.imread(path+p)[:,:,0]
    
    for i in range(3):
        img = img_orig[STARTY+i*GRID : STARTY+i*GRID+40, LEFT:LEFT+100]
        img_left.append(img)
        title_left.append(get_sum(img_orig[STARTY+i*GRID : STARTY+i*GRID+40, LEFT+50]))
    
    for i in range(3):
        img = img_orig[STARTY+i*GRID : STARTY+i*GRID+40, RIGHT:RIGHT+100]
        img_right.append(img)
        title_right.append(get_sum(img_orig[STARTY+i*GRID : STARTY+i*GRID+40, RIGHT+50]))
    
    fig, axs = plt.subplots(3, 2)
    axs[0][0].imshow(img_left[0])
    axs[0][0].set_title(title_left[0])
    axs[0][1].imshow(img_right[0])
    axs[0][1].set_title(title_left[1])
    axs[1][0].imshow(img_left[1])
    axs[1][0].set_title(title_left[2])
    axs[1][1].imshow(img_right[1])
    axs[1][1].set_title(title_right[0])
    axs[2][0].imshow(img_left[2])
    axs[2][0].set_title(title_right[1])
    axs[2][1].imshow(img_right[2])
    axs[2][1].set_title(title_right[2])
    plt.show()
    plt.close()
    
    if print_count < print_limit:
        break
    else:
        print_count += 1
```
